```python
class getSTD: 
    def __init__(self, filepath): 
        self.df = pd.read_excel(filepath)

    @property
    def getDF(self):
        A = self.df[['PDF_name'] +  list(self.df.columns[-3:])].groupby('PDF_name').std().dropna(how='any').reset_index()
        B = self.df[['PDF_name', "match_text_CRI"]].groupby("PDF_name").mean().dropna(how='any').reset_index()
        return pd.merge(pd.DataFrame(A),pd.DataFrame(B), on='PDF_name')
```

```python
filebundle = "/Users/improvise/Desktop/Guaranteed Graduates/Thesis/ESG/Playground/03_TextImgMatch/distance"
df_std = pd.DataFrame()
 
for year in range(2020,2023):
    for ReportType in ['ESG', 'SUS']:
        df_cache = getSTD(os.path.join(filebundle, "_".join([ReportType, str(year), 'distance.xlsx']))).getDF
        df_std = pd.concat([df_std, df_cache])
```

```python
df_std['Pub_year'] = df_std['PDF_name'].apply(lambda x: int(x.split("-")[-3]) - 1)
df_std['Corp_code'] =df_std['PDF_name'].apply(lambda x: x.split("-")[0][:6])
```

```python
df_std.rename(columns={'match_text_CRI_x':'match_text_CRI_std', 'match_text_CRI_y':'match_text_CRI_mean'}, inplace=True)
```

## 排序

```python
def getDeletionList(num=15, df_std=df_std):
    DeletionList=[]
    # match_text_CRI_mean 
    DeletionList = list(df_std.sort_values(by='match_text_CRI_mean', ascending=True).head(num)['Corp_code'])
    # match_text_prob_std 
    DeletionList = DeletionList + list(df_std.sort_values(by='match_text_prob', ascending=False).head(num)['Corp_code'])
    # distance
    DeletionList = DeletionList + list(df_std.sort_values(by='distance', ascending=False).head(num)['Corp_code'])
    # match_text_CRI_mean 
    DeletionList = DeletionList + list(df_std.sort_values(by='match_text_CRI_mean', ascending=False).head(num)['Corp_code'])

    return Counter(DeletionList)
```

```python
count = Counter(DeletionList)
print(count, len(count))
```

```python
DeletionList_15 = getDeletionList()
DeletionList_10 = getDeletionList(num = 10)
```
